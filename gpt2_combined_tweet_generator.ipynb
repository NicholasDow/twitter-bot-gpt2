```
import pandas as pd
tweets_of_interest = [
                      #'elonmusk',
                      'jehlpxyz',
                      #'kanyewest',
                      #'KimKardashian',
                      'LiangHanWei6',
                      'motivational',
                      'suigamsiM',
                      'TSM_Leffen',
                      #'officialmcafee',
                      'dril',
                      'KimKierkegaard',
                      'TheOnion'
                      ]
folder = 'drive/MyDrive/tweets/'

df = pd.DataFrame(columns = ['tweets'])
for name in tweets_of_interest:
  file = pd.read_csv(folder+name+'_tweets.csv')
  # This puts a limit on how many kim kardashian and motivational tweets I have to endure
  if len(file) > 5000:
    file = file[:5000]
  df = pd.concat([df, file])
df = df[df['tweets'].apply(lambda x: len(x) > 40 and len(x) < 500)]
df = df.sample(frac = 1)
file_name = 'combined'+'_tweets.csv'
df.to_csv(folder+file_name, index=False)
df.to_csv(file_name, index=False)
#gpt2.copy_file_from_gdrive('combined_tweets.csv')
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset='combined_tweets.csv',
              model_name='124M',
              steps=2000,
              restore_from='fresh',
              run_name='run2',
              print_every=10,
              sample_every=50,
              save_every=500
              )
```
